# Preprocessing

Steps in `src/preprocess.py`:
1. Load raw weather CSV; parse dates.
2. **Derived features (per location):** rolling 3d/7d precipitation, temperature mean, dry_days.
3. **Target:** `extreme_rain_tomorrow` = 1 if next day precipitation > 50 mm.
4. Drop rows with missing/infinite values; encode location as `location_id`.
5. Time-based split 70% / 15% / 15% → train, val, test.

```python
# Run full preprocessing (from project root: python -m src.preprocess)
# Here we load the already-processed outputs and inspect them.
import pandas as pd
import sys
from pathlib import Path
root = Path.cwd() if (Path.cwd() / "src").exists() else Path.cwd().parent
sys.path.insert(0, str(root))
from src.preprocess import load_raw, add_derived_features, add_target, get_feature_columns
from src.config import PROCESSED_DIR

# Load raw and apply steps for inspection (optional: skip if processed files exist)
raw = load_raw()
raw = add_derived_features(raw)
raw = add_target(raw)
print("After derived features + target, shape:", raw.shape)
print("Target value_counts:", raw['extreme_rain_tomorrow'].value_counts().to_dict())
```

```python
# Feature columns used for modeling
feature_cols = ['location_id'] + get_feature_columns()
print("Features:", feature_cols)
```

```python
# Load saved splits and show sizes
train = pd.read_csv(PROCESSED_DIR / 'train.csv')
val = pd.read_csv(PROCESSED_DIR / 'val.csv')
test = pd.read_csv(PROCESSED_DIR / 'test.csv')
print('Train:', len(train), 'Val:', len(val), 'Test:', len(test))
print('Target prevalence (train):', train['extreme_rain_tomorrow'].mean())
```
