*EDA - Exploratory Data Analysis*

Load raw data, inspect distributions, missing values, and basic statistics.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load raw weather (run fetch_weather first if missing)
df = pd.read_csv('../data/raw/weather_raw.csv')
df['date'] = pd.to_datetime(df['date'])
df.info()
df.describe()
df.head(10)
```

```python
# Rows per location and date range
df['location'].value_counts()
df[['date']].agg(['min', 'max'])
```

```python
# Missing values
df.isnull().sum()
```

```python
# Precipitation over time (one location)
loc = df['location'].iloc[0]
sns.lineplot(data=df[df['location']==loc], x='date', y='precipitation_sum', label=loc)
plt.xticks(rotation=45)
plt.ylabel('Precipitation (mm)')
plt.title('Daily precipitation sum')
plt.tight_layout()
plt.show()
```

```python
# Temperature distributions by location
sns.boxplot(data=df, x='location', y='temperature_2m_max')
plt.title('Max temperature by location')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
# Distribution of daily precipitation (zeros dominate)
df['precipitation_sum'].hist(bins=50, edgecolor='black', alpha=0.7)
plt.xlabel('Precipitation (mm)')
plt.ylabel('Count')
plt.title('Distribution of daily precipitation_sum')
plt.tight_layout()
plt.show()
```
